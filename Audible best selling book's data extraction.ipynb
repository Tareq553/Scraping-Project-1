# Library Insertion

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

# Main Code Portion

```python
#Checking the url
url = 'https://www.audible.com/adblbestsellers?page=1'
data = requests.get(url)
soup = BeautifulSoup(data.content, 'html.parser')
#print(soup.prettify())
```

```python
#Total Page Finding
total_page=soup.select_one('div.bc-col-responsive.bc-text-right.bc-col-8 > div > span > ul > li:nth-child(5) > a').text.strip()
total_page
```

```python
lst=[]
```

```python
for i in range(1,int(total_page)):
    print(f'Running page no--')
    url = f'https://www.audible.com/adblbestsellers?page={i}'
    data = requests.get(url)
    soup = BeautifulSoup(data.content, 'html.parser')
    main_field=soup.select_one("#product-list-a11y-skiplink-target")
    contents=main_field.select("li.bc-list-item.productListItem")
    for content in contents:
        try:
            book_name=content.select_one('li:nth-child(1) > h3 > a').text.strip()
        except:
            book_name='No book name'
        try:
            author_name=content.select_one('li.bc-list-item.authorLabel > span > a').text.strip()
        except:
            author_name='No author name'
        try:
            narrated_by=content.select_one('li.bc-list-item.narratorLabel > span > a').text.strip()
        except:
            narrated_by='No narrator name'
        try:
            length=content.select_one('li.bc-list-item.runtimeLabel > span').text.strip()
        except:
            length='No length mentioned'
        try:
            release_date=content.select_one('li.bc-list-item.releaseDateLabel > span').text.strip()
        except:
            release_date='No release date mentioned'
        try:
            language=content.select_one('li.bc-list-item.languageLabel > span').text.strip()
        except:
            language='No language mentioned'
        try:
            rating=content.select_one('li.bc-list-item.ratingsLabel > span.bc-text.bc-size-small.bc-color-secondary').text.strip()
        except:
            rating='No rating mentioned'
        try:
            price=content.select('span.bc-text.bc-size-base.bc-color-base')
        except:
            price='No price mentioned'
        f={
            'Book_name':book_name,
            'Author_name':author_name,
            'Narrated_by':narrated_by,
            'Length':length,
            'Release_date':release_date,
            'Language':language,
            'Rating':rating,
            'Price':price
        }
        lst.append(f)
        
            
        
    
```

# DataFrame Generation

```python
pd.DataFrame(lst)
```

# CSV Generation

```python
pd.DataFrame(lst).to_csv('Audible Best Selling Books.csv',index=False)
```
